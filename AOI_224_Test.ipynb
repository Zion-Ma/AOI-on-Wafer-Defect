```python
from pandas import read_csv
import pandas as pd
# load file as CSV
from sklearn.preprocessing import LabelBinarizer
import numpy
import numpy as np
filename = '/content/drive/MyDrive/2021-spring-programming lab/aoi/train.csv'
mapping1_csv = read_csv(filename,encoding="Big5")
filename = '/content/drive/MyDrive/2021-spring-programming lab/aoi/test.csv'
mapping2_csv = read_csv(filename,encoding="Big5")
mapping1_csv = mapping1_csv.values.tolist()
mapping2_csv = mapping2_csv.values.tolist()

import cv2
import os
from PIL import Image
from numpy import savez_compressed
from numpy import asarray

def load_dataset(directory,mapping_csv):
    X= list()
    data = []
    i=0
    # loop over the image paths
    for imagePath, label in mapping_csv:
        image = Image.open(directory + imagePath)
        image = image.convert('RGB')
        image = image.resize((224,224))
        image = asarray(image)/255
        data.append(image)
    # convert the data and labels to NumPy arrays while scaling the pixel
    # perform one-hot encoding on the labels
    from keras import utils as np_utils
    X.extend(data)
    return asarray(X)
```

```python
testX= load_dataset('test_images/',mapping2_csv[7000:])
print(testX.shape)
```

```python
from tensorflow import keras
model = keras.models.load_model('/content/drive/MyDrive/2021-spring-programming lab/AOI_Final_Report/model_Xception_11.h5')
```

```python
y_pred = model.predict(testX)
y_pred
```

```python
test_result=np.array([])
for i in range(len(y_pred)):
  test_result=np.append(test_result,np.where(y_pred[i]==np.max(y_pred[i]))[0]) # Find the index which has the highest probablity 
test_result=test_result.reshape(-1,1) # Transpose to fit in DataFrame format
test_result.shape
```

```python
file=pd.DataFrame(test_result)
file.head(10)
```
